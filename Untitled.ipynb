```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Данные загрязнения воздуха по годам для каждого города
data = {
    'Год': range(2011, 2020),
    'Алматы_загрязнение': [20, 21, 22, 23, 24, 25, 26, 27, 28],
    'Астана_загрязнение': [15, 16, 17, 18, 19, 20, 21, 22, 23],
    'Шымкент_загрязнение': [18, 19, 20, 21, 22, 23, 24, 25, 26],
    'Караганда_загрязнение': [12, 12.5, 13, 13.5, 14, 14.5, 15, 15.5, 16],
    'Атырау_загрязнение': [10, 10.5, 11, 11.5, 12, 12.5, 13, 13.5, 14],
}

# Создаем DataFrame
df = pd.DataFrame(data)

# Проверяем на наличие NaN
print("Проверка на NaN:")
print(df.isna().sum())

# Прогноз на 2020-2024
future_years = np.array(range(2020, 2025)).reshape(-1, 1)

# Функция для прогноза загрязнения
def forecast_pollution(city_data):
    model = LinearRegression()
    years = np.array(range(2011, 2020)).reshape(-1, 1)

    # Преобразуем city_data в NumPy массив и очищаем от NaN
    city_data = city_data.to_numpy().reshape(-1, 1)
    city_data = np.nan_to_num(city_data)  # Заменяем NaN на 0, если они есть

    # Проверка длины массивов
    if len(years) != len(city_data):
        raise ValueError(f"Несоответствие длин: years ({len(years)}) и city_data ({len(city_data)})")

    model.fit(years, city_data)
    forecast = model.predict(future_years)
    return forecast.flatten()  # Преобразуем обратно в одномерный массив

# Создаем прогноз для каждого города
results = []
for city in ['Алматы', 'Астана', 'Шымкент', 'Караганда', 'Атырау']:
    city_column = f'{city}_загрязнение'
    forecast = forecast_pollution(df[city_column])
    
    # Добавляем прогноз в DataFrame
    df_future = pd.DataFrame({
        'Год': future_years.flatten(),
        f'{city}_загрязнение': forecast
    })
    
    # Объединяем данные
    results.append(df_future)

# Объединяем все результаты в один DataFrame
df_results = pd.concat(results, axis=1)
df_results = df_results.loc[:, ~df_results.columns.duplicated()]  # Убираем дублирующиеся столбцы

# Выводим итоговый DataFrame с прогнозом
print("Итоговый DataFrame с прогнозом:")
print(df_results)

# График с прогнозом
plt.figure(figsize=(12, 8))
for city in ['Алматы', 'Астана', 'Шымкент', 'Караганда', 'Атырау']:
    plt.plot(df['Год'], df[f'{city}_загрязнение'], marker='o', label=city)

plt.title('Прогноз уровня загрязнения воздуха (2011-2024)')
plt.xlabel('Год')
plt.ylabel('Загрязнение воздуха (ед.)')
plt.axvline(x=2019, color='gray', linestyle='--', label='Начало прогноза')
plt.legend()
plt.grid()
plt.savefig('pollution_forecast.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Данные загрязнения воздуха по годам для каждого города
data = {
    'Год': range(2011, 2020),
    'Алматы_загрязнение': [20, 21, 22, 23, 24, 25, 26, 27, 28],
    'Астана_загрязнение': [15, 16, 17, 18, 19, 20, 21, 22, 23],
    'Шымкент_загрязнение': [18, 19, 20, 21, 22, 23, 24, 25, 26],
    'Караганда_загрязнение': [12, 12.5, 13, 13.5, 14, 14.5, 15, 15.5, 16],
    'Атырау_загрязнение': [10, 10.5, 11, 11.5, 12, 12.5, 13, 13.5, 14],
}

# Создаем DataFrame
df = pd.DataFrame(data)

# Проверяем на наличие NaN
print("Проверка на NaN:")
print(df.isna().sum())

# Прогноз на 2020-2021 годы
future_years = np.array(range(2020, 2022)).reshape(-1, 1)

# Функция для прогноза загрязнения
def forecast_pollution(city_data):
    model = LinearRegression()
    years = np.array(range(2011, 2020)).reshape(-1, 1)

    # Преобразуем city_data в NumPy массив и очищаем от NaN
    city_data = city_data.to_numpy().reshape(-1, 1)
    city_data = np.nan_to_num(city_data)  # Заменяем NaN на 0, если они есть

    # Проверка длины массивов
    if len(years) != len(city_data):
        raise ValueError(f"Несоответствие длин: years ({len(years)}) и city_data ({len(city_data)})")

    model.fit(years, city_data)
    forecast = model.predict(future_years)
    return forecast.flatten()  # Преобразуем обратно в одномерный массив

# Создаем прогноз для каждого города
results = []
for city in ['Алматы', 'Астана', 'Шымкент', 'Караганда', 'Атырау']:
    city_column = f'{city}_загрязнение'
    forecast = forecast_pollution(df[city_column])
    
    # Добавляем прогноз в DataFrame
    df_future = pd.DataFrame({
        'Год': future_years.flatten(),
        f'{city}_загрязнение': forecast
    })
    
    # Объединяем данные
    results.append(df_future)

# Объединяем все результаты в один DataFrame
df_results = pd.concat(results, axis=1)
df_results = df_results.loc[:, ~df_results.columns.duplicated()]  # Убираем дублирующиеся столбцы

# Создаем реальные данные загрязнения на 2020-2021 годы
real_data = {
    'Год': [2020, 2021],
    'Алматы_загрязнение': [30, 32],  # Пример реальных данных
    'Астана_загрязнение': [25, 28],
    'Шымкент_загрязнение': [26, 29],
    'Караганда_загрязнение': [17, 18],
    'Атырау_загрязнение': [14, 15],
}

# Создаем DataFrame с реальными данными
df_real = pd.DataFrame(real_data)

# Объединяем прогноз и реальные данные для сравнения
df_comparison = pd.merge(df_results, df_real, on='Год', suffixes=('_Прогноз', '_Реальные'))

# Выводим итоговый DataFrame с прогнозом и реальными данными
print("Итоговый DataFrame с прогнозом и реальными данными:")
print(df_comparison)

# Вычисляем ошибки прогноза
errors = {}
for city in ['Алматы', 'Астана', 'Шымкент', 'Караганда', 'Атырау']:
    errors[city] = df_comparison[f'{city}_загрязнение_Прогноз'] - df_comparison[f'{city}_загрязнение_Реальные']

# Добавляем ошибки в DataFrame
for city, error in errors.items():
    df_comparison[f'{city}_Ошибка'] = error

# График с прогнозом и реальными данными
plt.figure(figsize=(12, 8))
for city in ['Алматы', 'Астана', 'Шымкент', 'Караганда', 'Атырау']:
    plt.plot(df_comparison['Год'], df_comparison[f'{city}_загрязнение_Прогноз'], marker='o', label=f'Прогноз {city}')
    plt.plot(df_comparison['Год'], df_comparison[f'{city}_загрязнение_Реальные'], marker='x', linestyle='--', label=f'Реальные {city}')

plt.title('Сравнение прогноза и реальных данных уровня загрязнения воздуха (2020-2021)')
plt.xlabel('Год')
plt.ylabel('Загрязнение воздуха (ед.)')
plt.axvline(x=2019, color='gray', linestyle='--', label='Начало прогноза')
plt.legend()
plt.grid()
plt.savefig('pollution_forecast_comparison.png')
plt.show()
```
